```python
import os
import pandas as pd
import datetime as dt
```

```python
input_data_path = '../data'
input_data_filename = 'binance_BTCUSDT_15m_from_2020_01_01_to_2021_12_31_candlesticks_signals_processed_technical_indicators_and_crypto_index'
input_data_extension = ".csv"
full_path_input_data = os.path.join(input_data_path, input_data_filename + input_data_extension)
```

```python
df = pd.read_csv(full_path_input_data)
```

```python
df.head()
```

```python
df['datetime'] = df.open_time.apply(lambda x: dt.datetime.fromtimestamp(x / 1000.0))
```

```python
df.head()
```

```python
date_value = df['datetime'].apply(lambda x: x.strftime('%Y-%m-%d %H:%M:%S'))
df.insert(0, 'time', date_value)
```

```python
df.head()
```

```python
df[df.duplicated(['time'])]
```

```python
df = df.drop_duplicates('time')
```

```python
export_path = "../data"
export_filename = 'binance_BTCUSDT_15m_from_2020_01_01_to_2021_12_31_candlesticks_signals_processed_technical_indicators_and_crypto_index_v2'
export_extension = ".csv"
full_export_path = os.path.join(export_path, export_filename + export_extension)
```

```python
df.to_csv(full_export_path, index=False)
```
